```python
inventaire = {
    "babanes": 120,
    "mangues":85,
    "ananas":45,
    "noix de coco":60,
    "papayes":30
}
```

```python
def afficher_inventaire (inventaire):
    print("Inventaire actuel de la plantation :")
    for fruit, quantile in inventaire.items():
        print(f"- {fruit.capitalize()} : {quantile} unités")
```

```python
afficher_inventaire(inventaire)
```
