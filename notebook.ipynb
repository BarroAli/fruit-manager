```python
inventaire = {
    "bananes": 120,
    "mangues":85,
    "ananas":45,
    "noix de coco":60,
    "papayes":30
}
```

```python
def afficher_inventaire (inventaire):
    print("Inventaire actuel de la plantation :")
    for fruit, quantile in inventaire.items():
        print(f"- {fruit.capitalize()} : {quantile} unités")
```

```python
def recolter (inventaire, fruit, quantite):
    inventaire[fruit] = inventaire.get(fruit, 0) + quantite
    print(f"\n Récolté {quantite} {fruit} supplémentaies !")
```

```python
recolter(inventaire, fruit="bananes", quantite=10)
```

```python
afficher_inventaire(inventaire)
```
